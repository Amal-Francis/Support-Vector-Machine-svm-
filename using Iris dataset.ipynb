```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['flower names'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'] , df0['sepal width (cm)'], color='g', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='b', marker='.')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='r', marker='*')
plt.xlabel('Sepal length (cm)', color='#40020e', fontweight='bold')
plt.ylabel('Sepal width (cm)', color='#40020e', fontweight='bold')
plt.title('Sepal Length vs Sepal Width', color='darkblue', fontweight='bold', fontsize=15)
plt.legend(iris.target_names)
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='g', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='b', marker='.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='r', marker='*')
plt.xlabel('Petal Length (cm)', color='#40020e', fontweight='bold')
plt.ylabel('Petal Width (cm)', color='#40020e', fontweight='bold')
plt.title('Petal Length vs Petal Width', color='darkblue', fontweight='bold', fontsize=15)
plt.legend(iris.target_names)
plt.show()
```

```python
plt.scatter(df0['sepal length (cm)'], df0['petal length (cm)'], color='g', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['petal length (cm)'], color='b', marker='.')
plt.scatter(df2['sepal length (cm)'], df2['petal length (cm)'], color='r', marker='*')
plt.xlabel('Sepal Length (cm)', color='#40020e', fontweight='bold')
plt.ylabel('Petal Length (cm)', color='#40020e', fontweight='bold')
plt.title('Sepal Length vs Petal Length', fontweight='bold', color='darkblue', fontsize=15)
plt.legend(iris.target_names)
plt.show()
```

```python
plt.scatter(df0['sepal width (cm)'], df0['petal width (cm)'], color='g', marker='+')
plt.scatter(df1['sepal width (cm)'], df1['petal width (cm)'], color='b', marker='.')
plt.scatter(df2['sepal width (cm)'], df2['petal width (cm)'], color='r', marker='*')
plt.xlabel('Sepal width (cm)', color='#40020e', fontweight='bold')
plt.ylabel('Petal width (cm)', color='#40020e', fontweight='bold')
plt.title('Sepal Width vs Petal Width', color='darkblue', fontweight='bold', size=15)
plt.show()
```

```python
x=df.drop(['target', 'flower names'], axis=1)
y = df.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = SVC(C=1000)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
conf = confusion_matrix(y_test, y_pred)
sns.heatmap(conf, annot=True, cmap='viridis')
```

```python
p_df=pd.DataFrame({'Original Value of Y': y_test,
             'Predicted Value':  y_pred}, index=None)
p_df.reset_index(drop=True, inplace=True)
p_df
```
